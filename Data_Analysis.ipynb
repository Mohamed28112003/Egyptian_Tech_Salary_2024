```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px

```

```python
df = pd.read_csv('/content/Tech_Egyptian_Salary.csv')
```

```python
df.head()
```

# Count Jop Titles

```python


plt.figure(figsize=(15, 8))

sns.set_style("darkgrid")

palette = sns.color_palette("flare", n_colors=df['Title'].nunique())

sns.countplot(data=df, y='Title', order=df['Title'].value_counts().index, hue='Title', palette=palette, dodge=False, legend=False)

plt.title('Count of Job Titles')
plt.xlabel('Count')
plt.ylabel('Job Title')

plt.show()
```

# Salary Distribution For Each Job Title

```python
plt.figure(figsize=(16, 8))

sns.barplot(data=df, y='Salary', x='Title', hue='Title', palette='flare', dodge=False, legend=False,errorbar=None)

plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary (EGP)')
plt.xticks(rotation=90)
plt.show()
```

# Compare Each Jop Title By Salary



* I creat function to compare jop titles by salaray according to years of experiences 



```python
def plot_salary_years(df, job_titles):
  filtered_df = df[df['Title'].isin(job_titles)]

  plt.figure(figsize=(20, 10))

  sns.barplot(data=filtered_df, x='Years of Experiences ', y='Salary', hue='Title', palette='rocket',errorbar=None)
  title = ' vs '.join(job_titles)
  plt.title(f'{title}')
  plt.ylabel('Salary (EGP)')
  #plt.xticks(rotation=90)
  plt.show()



```



* I creat onther function to compare jop titles by salaray only 



```python


def plot_salary_title(df, job_titles):
    filtered_df = df[df['Title'].isin(job_titles)]

    plt.figure(figsize=(20, 10))

    sns.barplot(data=filtered_df, x='Title', y='Salary', hue='Title', palette='flare', dodge=False, legend=False,errorbar=None)

    title = ' vs '.join(job_titles)
    plt.title(f'{title}')
    plt.ylabel('Salary (EGP)')
    plt.xlabel('Job Title')

    plt.show()




```



* Compare web developer 



```python
job_titles = ['Frontend Developer', 'Backend Developer', 'Full Stack Developer']

plot_salary_years(df, job_titles)
```

```python
plot_salary_title(df, job_titles)
```



* Compare frameworks for web developer 



```python
job_titles = ['.Net Developer', 'PHP Developer']
plot_salary_years(df, job_titles)
```

```python
plot_salary_title(df, job_titles)
```




*  compare data jops 





```python
job_titles = ['Data Analyst', 'Data Scientist','Data Engineer','Business Intelligence Engineer']
plot_salary_years(df, job_titles)
```

```python
plot_salary_title(df, job_titles)
```





* compare Mobile Development 





```python
job_titles = ['Flutter Developer', 'Android Developer','IOS Developer']
plot_salary_years(df, job_titles)
```

```python
plot_salary_title(df, job_titles)
```

# Work Type

```python

work_type_counts = df['Work Type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(work_type_counts, labels=work_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Work Type Distribution')
plt.show()
```

```python


plt.figure(figsize=(15, 8))

sns.barplot(data=df, y='Salary', x='Work Type', hue='Work Type', palette='flare', dodge=False, legend=False,errorbar=None)
plt.title('Salary Distribution by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Salary (EGP)')

plt.show()
```

# City of Company site

```python


plt.figure(figsize=(12, 8))

palette = sns.color_palette("flare", n_colors=df['City of Company site'].nunique())

sns.countplot(data=df, y='City of Company site', order=df['City of Company site'].value_counts().index, hue='City of Company site', palette=palette, dodge=False, legend=False)

plt.title('Count of City of Company site')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```
